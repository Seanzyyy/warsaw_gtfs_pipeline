```python
from pyspark.sql.functions import to_date, col
```

# 1. Load calendar_dates table from bronze

```python
calendar_dates_df = spark.table("warsaw_gtfs.bronze.calendar_dates")
calendar_dates_df.show(50)

# data issues
# 1. Not friendly field names
# 2. Service ID with dat in some rows
# 3. Date field not proper data type.
```

#  2. Clean data


```python
# 1. Rename columns
calendar_dates_df = calendar_dates_df.withColumnRenamed("date", "service_date").withColumnRenamed("exception_type", "is_exception")

# 2. Format service_date to date type

calendar_dates_df = calendar_dates_df.withColumn("service_date", to_date("service_date", "yyyyMMdd"))


# 3. Remove date from service_id
calendar_dates_df = calendar_dates_df.withColumn("service_id", col("service_id").substr(-3,3))

calendar_dates_df.show(50)

```

# 3. Load data

```python
calendar_dates_df.write.mode("overwrite").format("delta").saveAsTable("warsaw_gtfs.silver.calendar_dates")
```
