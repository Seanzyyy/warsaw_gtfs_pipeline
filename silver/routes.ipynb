```python
from pyspark.sql import functions as F
```

#1. Load bronze.routes data


```python
df = spark.read.table('warsaw_gtfs.bronze.routes')
df.display()
```

##EDA

```python
# 1. Route types- 0 = tram routes, 1 = metro routes, 2 = skm routes, 3 = bus routes.
```

#2. Transformation


```python
df = df.withColumn(
    'route_type', 
    F.when(F.col('route_type') == 0, 'tram')
    .when(F.col('route_type') == 1, 'metro')
    .when(F.col('route_type') == 2, 'skm')
    .when(F.col('route_type') == 3, 'bus')
    .otherwise('other')
)
```

#3. Write routes table to silver

```python
df.write.mode('overwrite').saveAsTable('warsaw_gtfs.silver.routes')
```
