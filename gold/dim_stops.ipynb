```python
%sql
CREATE OR REPLACE TABLE warsaw_gtfs.gold.dim_stops AS
SELECT 
stop_id,
zone_id,
CONCAT(stop_name, ' ', stop_code) AS stop_name,
stop_name_stem AS stop_group,
platform_code AS platform_code_for_metro,
location_type AS stop_type,
parent_station,
wheelchair_boarding,
town_name,
street_name
FROM warsaw_gtfs.silver.stops
```
