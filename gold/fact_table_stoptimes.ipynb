```python
%sql
CREATE OR REPLACE TABLE warsaw_gtfs.gold.fact_stop_times AS
SELECT 
st.trip_id,
t.route_id,
st.stop_id,
t.service_id,
st.stop_sequence,
st.pickup_type,
st.arrival_time,
st.departure_time
FROM warsaw_gtfs.silver.stop_times st 
INNER JOIN warsaw_gtfs.silver.trips  t ON st.trip_id = t.trip_id
```
