#1. Extract data from url source

```python
import requests
from datetime import datetime

# source url
url = 'https://mkuran.pl/gtfs/warsaw/alerts.pb'

# destination path
volume_path = '/Volumes/warsaw_gtfs/default/text_data/'

# generate unique filename (timestamped)
filename = f"gtfs_rt_{datetime.utcnow().strftime('%Y%m%d_%H%M%S')}.pb"

full_path = volume_path + filename

# fetch data
response = requests.get(url)
response.raise_for_status()

# write binary file
with open(full_path, "wb") as f:
    f.write(response.content)

print(f"Saved raw protobuf to: {full_path}")
```
